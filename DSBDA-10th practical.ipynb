```python
import seaborn as sns
import pandas as pd
```

```python
iris = sns.load_dataset('iris')
print(iris.dtypes)
```

```python
import matplotlib.pyplot as plt
iris.hist(figsize=(10, 8), bins=20, color='skyblue', edgecolor='black')
plt.suptitle('Histogram of Iris Features', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=iris)
plt.title('Boxplot of All Iris Features')
plt.show()
```

```python
for column in iris.select_dtypes(include='float'):
    plt.figure(figsize=(8, 4))
    sns.boxplot(x='species', y=column, data=iris)
    plt.title(f'Boxplot of {column.capitalize()} by Species')
    plt.show()
```
